### Data->Array

```python
x=df.values
```

```python
x[0]
```

```python

ordi=OrdinalEncoder()

os=OneHotEncoder(drop="first")

sc=StandardScaler()
```

```python
from sklearn.compose import ColumnTransformer
trf = ColumnTransformer([
        ("OneHot",os,[0,2,3,5,6,8,9,10,11,12,13,15]),
        ('ordi',ordi,[7,14,16]),
        ("stnader",sc,[4,17,18])
    ],remainder='passthrough')
```

```python
x=trf.fit_transform(x)
```

```python
x[0]
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
km = KMeans(n_clusters=2)
labels = km.fit_predict(x)
```

```python
labels
```

```python
y=NormalData['Churn'].values
```

```python
y
```

```python
t=0
total=len(y)
for i in range(len(y)):
    if labels[i]==y[i]:
        t+=1
    else:
        pass
    
```

```python
print(f"True predicted={t} out of {total}")
```

```python
print(f"%={(t/total)*100} ")
```
